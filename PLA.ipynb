```python
def vadd(v,w):
    return [v[i]+w[i] for i in range(len(v))]
def vmul(s,v):
    return [s*v[i] for i in range(len(v))]
```

```python
w= [1,1,1]
plt.scatter(x1,y1, color = 'r')
plt.scatter(x2,y2, color = 'b')
p1,p2=plotline(w,[-10,10],[10,10])
plt.plot([p1[0],p2[0]],[p1[1],p2[1]])
plt.xlim(-5,10)
plt.ylim(-5,10)
axes=plt.gca()
axes.set_aspect(1)
```
